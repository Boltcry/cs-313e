```python
def quick_sort(array):
    if len(array) < 2:
        return array
    else:
        pivot = array[0]
        less = [i for i in array[1:] if i <= pivot]
        greater = [i for i in array[1:] if i > pivot]
        return quick_sort(less) + [pivot] + quick_sort(greater)

quick_sort([4, 3, 2, 1])
```

```python
def n_fibonacci(n: int) -> int:
    if n == 0 or n == 1:
        return n

    return n_fibonacci(n - 1) + n_fibonacci(n - 2)

n_fibonacci(1)
```

```python
def find_strong_passwords(arr: list[str]) -> None:
    A_ASCII = ord('a')
    Z_ASCII = ord('z')

    ZERO_ASCII = ord('0')
    NINE_ASCII = ord('9')

    for i in range(len(arr)):

        if len(arr[i]) < 5:
            continue

        digit_count = 0
        letter_count = 0

        for j in range(len(arr[i])):

            if A_ASCII <= ord(arr[i][j].lower()) <= Z_ASCII:
                letter_count += 1

            elif ZERO_ASCII <= ord(arr[i][j]) <= NINE_ASCII:
                digit_count += 1
            
            if letter_count == len(arr[i]) or digit_count == len(arr[i]):
                continue
                
            print(arr[i], end=' ')
```
